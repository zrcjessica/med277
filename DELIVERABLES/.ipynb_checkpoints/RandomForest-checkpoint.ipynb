Preprocess

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize, sent_tokenize, TweetTokenizer 
from nltk.corpus import stopwords
from string import punctuation
import re
```

```python
# read in datasets
train_original = pd.read_csv('train.csv')
train = pd.read_csv('train_medical.csv')
```

```python
# remove english stopwords
def remove_stopwords(words):
    stop_words = set(stopwords.words('english'))
    return [word.lower() for word in words if word not in stop_words]

# remove punctuation
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

# preprocess and balance dataset
def Preprocess(train):
    sample_size = len(train[train.target == 1])
    train_balanced = train[train.target == 1].sample(sample_size).append(train[train.target == 0].sample(sample_size)).reset_index()
    train_balanced = train_balanced.drop(columns=['index'])
    # Removing punctuation
    train_balanced['question_text_token'] = train_balanced['question_text'].apply(lambda x: remove_punctuation(x))
    # Tokenizing the text
    train_balanced['question_text_token'] = train_balanced['question_text_token'].apply(lambda x: word_tokenize(x))
    # Removing stopwords
    train_balanced['question_text_token'] = train_balanced['question_text_token'].apply(lambda x: remove_stopwords(x))
    return train_balanced
```

```python
train_balanced_original = Preprocess(train_original)
train_balanced = Preprocess(train)
```

Google Word Vectors <br>
Please download Google News word2vec embeddings from [Kaggle](https://www.kaggle.com/c/quora-insincere-questions-classification/data) to proceed. This is a large file! 

```python
# google news word2vec model
google_word2vec = 'GoogleNews-vectors-negative300.bin'
```

```python
from gensim.models.word2vec import Word2Vec
from gensim.models import KeyedVectors
# load word2vec model
model = KeyedVectors.load_word2vec_format(google_word2vec, binary=True)
```

```python
# transform questions to average vector representation
def AverageVectorization(question_token):
    words = [i for i in question_token if i in model]
    vector_representations = [model[i] for i in words]
    vector_representation = [float(sum(col))/len(col) for col in zip(*vector_representations)]
    return vector_representation
```

```python
# transform original data
train_balanced_original['question_vector_avg'] = train_balanced_original['question_text_token'].apply(AverageVectorization)
```

```python
#transform filtered data
train_balanced['question_vector_avg'] = train_balanced['question_text_token'].apply(AverageVectorization)
```

```python
# drop questions that could not be transformed properly
train_balanced_original = train_balanced_original[train_balanced_original['question_vector_avg'].apply(lambda x: len(x) == 300)]
```

PCA

```python
from sklearn.decomposition import PCA
```

```python
# PCA for balanced data
def PreparePCA(train_balanced):
    y = np.array(train_balanced['target'])
    X = list(train_balanced['question_vector_avg'])
    pca = PCA(n_components=2, svd_solver='full')
    X_pca = pca.fit_transform(X)
    Y = X_pca
    y = np.array(train_balanced['target'])
    return Y, y
```

```python
Y, y = PreparePCA(train_balanced_original)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('PCA for Original Data')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')

for lab, col in zip((0, 1),
                        ('red', 'green')):
        ax.scatter(Y[y==lab, 0],
                    Y[y==lab, 1],
                    label=lab,
                    c=col)
plt.savefig('original_pca.png')
```

```python
Y, y = PreparePCA(train_balanced)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('PCA for Filtered Data')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')

for lab, col in zip((0, 1),
                        ('red', 'green')):
        ax.scatter(Y[y==lab, 0],
                    Y[y==lab, 1],
                    label=lab,
                    c=col)
plt.savefig('filtered_pca.png')
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.externals import joblib
```

```python
# split data 50/50 for training and testing & save .pkl for testing

# train on balanced dataset for filtered biomedical dataset
train_balanced = train_balanced.sample(frac=1).reset_index()
train_balanced = train_balanced.drop(columns=['index'])
train_balanced_train = train_balanced.iloc[:int(0.5*len(train_balanced))]
train_balanced_test = train_balanced.iloc[int(0.5*len(train_balanced)):]
joblib.dump(train_balanced_test, 'random_forest_test.pkl')

# train on balanced dataset for unfiltered dataset
train_balanced_original = train_balanced_original.sample(frac=1).reset_index()
train_balanced_original = train_balanced_original.drop(columns=['index'])
train_balanced_original_train = train_balanced_original.iloc[:int(0.5*len(train_balanced_original))]
train_balanced_original_test = train_balanced_original.iloc[int(0.5*len(train_balanced_original)):]
joblib.dump(train_balanced_original_test, 'random_forest_original_test.pkl')
```

```python
# Fit a random forest to the training data for unfiltered data, using 100 trees
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(list(train_balanced_original_train['question_vector_avg']), list(train_balanced_original_train['target']))
```

```python
joblib.dump(forest, 'RandomForestBalanced_original.pkl')
```

```python
# Fit a random forest to the training data for filtered data, using 100 trees
forest = RandomForestClassifier(n_estimators = 100)
forest = forest.fit(list(train_balanced_train['question_vector_avg']), list(train_balanced_train['target']))
```

```python
joblib.dump(forest, 'RandomForestBalanced.pkl')
```
